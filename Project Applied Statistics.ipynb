## Question :- An electrical manufacturing company conducts quality checks at specified periods on the products it manufactures. Historically,the failure rate for the manufactured item is 5%. Suppose a random sample of 10 manufactured items is selected.Answer the following questions.

```python
# It is a binomial distribution
```

```python
#Failure rate for manufactured item is 5%
#size of random sample is 10

#import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

import scipy.stats as stats
```

```python
#declare sample size in variable n
n=10

#declare p which represents the probability of failure rate
p=0.05

#declare a numpy array  with different possible number of manufactured items selected
k = np.arange(0,11)
k
```

```python
#import the required function
from scipy.stats import binom

binomial = binom.pmf(k=k, n=n, p=p)
binomial
```

# 2.A. Probability that none of the items are defective?

```python
#Probability that none of the items are defective
binomial[0]
```

###### Probability that none of the items are defective is 0.5987

# 2.B. Probability that exactly one of the items is defective?

```python
#Probability that exactly one of the items is defective
binomial[1]
```

###### Probability that exactly one of the items is defective is 0.3151

# 2.C. Probability that two or fewer of the items are defective?

```python
#Probability that two or fewer of the items are defective
binom.cdf(k=2, n=n,p=p)
```

###### Probability that two or fewer of the items are defective is 0.9885

# 2.D. Probability that three or more of the items are defective ?

```python
#Probability that three or more of the items are defective 
1-binom.cdf(k=2, n=n,p=p)
```

######  Probability that three or more of the items are defective is 0.0115

## Question :- A car salesman sells on an average 3 cars per week.

```python
#car salesman sells on an average 3 cars per week

#import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

import scipy.stats as stats


rate = 3

#declaring an array
n=np.arange(0,20)
n
```

```python
#constructing the poisson distribution
poisson = stats.poisson.pmf(n,rate)
poisson
```

```python
cumulative_poisson_dist = stats.poisson.cdf(n,rate)
```

# 3.A. What is Probability that in a given week he will sell some cars?

```python
#Probability that in a given week he will sell some cars
#P(X>0) = 1-P(X=0)
1-poisson[0]
```

###### Probability that in a given week car salesman will sell some cars is 0.9502

# 3.B. What is Probability that in a given week he will sell 2 or more but less than 5 cars?

```python
#Probability that in a given week he will sell 2 or more but less than 5 cars
#P(2≤X<5)
#P(x=2)+p(x=3)+p(x=4)
poisson[2]+poisson[3]+poisson[4]
```

###### Probability that in a given week he will sell 2 or more but less than 5 cars is 0.6161

# 3.C. Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week.

```python
#Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week
plt.plot(n,cumulative_poisson_dist, 'o-')
plt.title('Poisson distribution')
plt.xlabel('Number of cars sold per week')
plt.ylabel('Cumulative Probability of cars sold per week')
plt.show()
```

## Question : - Accuracy in understanding orders for a speech based bot at a restaurant is important for the Company X which has designed,marketed and launched the product for a contactless delivery due to the COVID-19 pandemic. Recognition accuracy that measures thepercentage of orders that are taken correctly is 86.8%. Suppose that you place an order with the bot and two friends of yoursindependently place orders with the same bot. Answer the following questions.

```python
#Recognition accuracy that measures the percentage of orders that are taken correctly is 86.8%
#size of random sample is 3

#import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

import scipy.stats as stats
```

```python
#declare sample size in variable n
n=3

#declare p which represents the probability of success rate
p=0.868

#declare a numpy array  with different possible number of manufactured items selected
k = np.arange(0,4)
k
```

```python
#import the required function
from scipy.stats import binom

binomial = binom.pmf(k=k, n=n, p=p)
binomial
```

# 4.A. What is the probability that all three orders will be recognised correctly?

```python
#probability that all three orders will be recognised correctly
binomial[3]
```

###### probability that all three orders will be recognised correctly is 0.6540

# 4.B. What is the probability that none of the three orders will be recognised correctly?

```python
#probability that none of the three orders will be recognised correctly
binomial[0]
```

# 4.C. What is the probability that at least two of the three orders will be recognised correctly?

```python
#probability that at least two of the three orders will be recognised correctly
#p(x>=2)
#1-p(x<1)
#calculate cdf
1-binom.cdf(k=1,n=n,p=p)
```
